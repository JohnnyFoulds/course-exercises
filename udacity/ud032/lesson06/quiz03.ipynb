```python
c = 0

def is_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def get_list_area(value):
    value = value.replace("{", "").replace("}", "")
    values = value.split("|")
    
    if len(values) > 2:
        raise Exception("There are more than 2 values in the list")
    else:
        first = float(values[0])
        second = float(values[1])
        
        area = first if len(str(first/ 1000000.0)) > len(str(second/ 1000000.0)) else second
        
    return area
    
def fix_area(area):
    if is_float(area):
        area = float(area)
    elif area.startswith("{"):
        area = get_list_area(area)
    else:
        area = None
        
    # show the first 10 rows
    global c
    if area != None:
        if c < 10:
            print(area)
            c += 1

    return area

if __name__ == "__main__":
    test()
```
