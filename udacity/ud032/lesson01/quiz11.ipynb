```python
#!/usr/bin/env python

import xlrd
from zipfile import ZipFile
datafile = "2013_ERCOT_Hourly_Load_Data.xls"


def open_zip(datafile):
    with ZipFile('{0}.zip'.format(datafile), 'r') as myzip:
        myzip.extractall()
        
def get_times(sheet, col, nrows):
    times = []
    hour_end = sheet.col_values(col, start_rowx=1, end_rowx=nrows-1)
    for dt in hour_end:
        times.append(xlrd.xldate_as_tuple(dt, 0))
        
    return times


def parse_file(datafile):
    workbook = xlrd.open_workbook(datafile)
    sheet = workbook.sheet_by_index(0)

    # get a list of time values
    times = get_times(sheet, 0, sheet.nrows)
    coast = sheet.col_values(1, start_rowx=1, end_rowx=sheet.nrows)
    
    data = {
            'maxtime': times[coast.index(max(coast))],
            'maxvalue': max(coast),
            'mintime': times[coast.index(min(coast))],
            'minvalue': min(coast),
            'avgcoast': sum(coast) / len(coast)
    }
    return data


def test():
    open_zip(datafile)
    data = parse_file(datafile)

    assert data['maxtime'] == (2013, 8, 13, 17, 0, 0)
    assert round(data['maxvalue'], 10) == round(18779.02551, 10)


test()
```
